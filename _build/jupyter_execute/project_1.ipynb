```python
import csv

data = []
indicator_name = "Adolescent fertility rate (births per 1,000 women ages 15-19)"
with open('/Users/josefinadesolminihac/Desktop/computing/GenderIndicators.csv', mode='r', encoding='utf-8-sig') as file:
    reader = csv.reader(file)
    
    # Skip the first 4 rows because they have the title or are empty
    for _ in range(4):
        next(reader)
    
    reader = csv.DictReader(file)
    
    # Filter the data based on the indicator name and collect 2022 values
    for row in reader:
        if row.get('Indicator Name') == indicator_name:
            value_str = row['2022']
            if value_str != '':   #had a problem when changing the type to float with the " " value, so I need to skip it
                value = float(value_str)
                data.append(value)

data
```

```python
from statistics import mean, median, multimode
mean_data = mean(data)
print(mean_data)
```

The median adolescent fertility rate is 33.27 births per 1,000 women aged 15-19. This suggests that half of the values in the dataset are below 33.27, and the other half are above, showing the central point of the data distribution.

```python
mode_data = multimode(data)
print(mode_data)
```

The modes of the adolescent fertility rate are 27.85 and 98.72 births per 1,000 women aged 15-19. These values represent the most frequently occurring fertility rates in the dataset, indicating that these rates are observed more often than other values.

```python
#For the visualization, I want to group the countries by quartile. 
import numpy as np

data = filtered_df_2022['2022']
#Find the quartiles
q1 = np.quantile(data, 0.25)
q2 = np.quantile(data, 0.5)
q3 = np.quantile(data, 0.75)
print(q1,q2,q3)
```

```python
#Create a function that assigns the quartile
def assign_quartile(row):
    if row["2022"] <= q1:
        return "Q1"
    elif q1 < row["2022"] <= q2:
        return "Q2"
    elif q2 < row["2022"] <= q3:
        return "Q3"
    else:
        return "Q4"
```

```python
# Apply the function to the DataFrame
filtered_df_2022["Quartile"] = filtered_df_2022.apply(assign_quartile, axis=1)
filtered_df_2022
```

```python
quartile_data = filtered_df_2022.groupby(['Quartile'], as_index=False)['2022'].mean().round(2)
quartile_data
```

```python
# Calculate the max value in the "2022" column to use as the reference for scaling
max_value = quartile_data["2022"].max()

# Define a function to create the sparkline scale with '*', using a scale of 10 (10 '*' is the biggest number)
def create_sparkline(value, max_value, scale_factor=10):
    num_stars = int((value / max_value) * scale_factor)
    return "*" * num_stars

# Apply the function to each value in the "2022" column
quartile_data["sparkline"] = quartile_data["2022"].apply(create_sparkline, max_value=max_value)

# Display the resulting DataFrame
quartile_data
```

```python
#create the table
print(f'{'Quartile':<20} {'Adolescent fertility rate (births per 1,000 women ages 15-19)':<65}')
print('_'*18+ ' ' + '_'*80)
print(f'{'':<20} {'Adolescent fertility rate (births per 1,000 women ages 15-19)':<65} {'Sparkline':<65}')
for index, row in quartile_data.iterrows():
    print(f'{row["Quartile"]:<20} {row["2022"]:<65} {row["sparkline"]:<30}')
```
